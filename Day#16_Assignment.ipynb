```python
## Import libraries to read dataset named Loan Default
## Import proper statistical library
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statistics
```

```python
## Calculate Mean, Median, Mode, and Standard Deviation of the amount of Loan disbursed in the column named disbursed_amount.
df = pd.read_csv('Loan_Defaults.csv')
print("mean of amount of Loan disbursed: "+str(statistics.median(df['DISBURSED_AMOUNT'])))
print('median of amount of Loan disbursed:'+str(statistics.mean(df['DISBURSED_AMOUNT'])))
print('mode of amount of Loan disbursed: '+str(statistics.mode(df['DISBURSED_AMOUNT'])))
print('Standard Deviation of amount of Loan disbursed: '+str(statistics.stdev(df['DISBURSED_AMOUNT'])))
```
